3. Graph the county with the highest population and the corresponding state from - https://raw.githubusercontent.com/plotly/datasets/master/minoritymajority.csv . Can you make the county within the state a specific color?

```python
pop_df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/minoritymajority.csv')
```

```python
top_df = pop_df.nlargest(1, 'TOT_POP')
top_df.head(1)
```

```python
scope = ['California']
#df name
df_new = pop_df[pop_df['STNAME'].isin(scope)]

# list values
values = df_new['TOT_POP'].tolist()
fips = df_new['FIPS'].tolist()

bupu = cl.scales['9']['seq']['BuPu']
bupu500 = cl.interp( bupu, 65 )
fig = ff.create_choropleth(
    fips=fips, values=values, scope=scope, 
    colorscale=bupu500, round_legend_values=True,
    simplify_county=0, simplify_state=0,
    county_outline={'color': 'rgb(15, 15, 55)', 'width': 0.5},
    state_outline={'width': 0.5},
    legend_title='Population per County',
    title='California')

py.iplot(fig, filename='California Choropleth')
# IFrame(fig, width=700, height=350)
```

Ranges and Highest Pop County
